```python
# plots
import matplotlib.pyplot as plt
import numpy as np
plt.rcParams["font.size"] = 18
from matplotlib import cycler
colors = cycler('color',
                ['#EE6666', '#3388BB', '#9988DD',
                 '#EECC55', '#88BB44', '#FFBBBB'])
plt.rc('axes', facecolor='#E6E6E6', edgecolor='none',
       axisbelow=True, grid=False, prop_cycle=colors)
plt.rc('grid', color='w', linestyle='solid')
plt.rc('xtick', direction='out', color='black', labelsize=16)
plt.rc('ytick', direction='out', color='black', labelsize=16)
font = {'family' : 'normal',
        'size'   : 18}
plt.rc('font', **font)
plt.rc('patch', edgecolor='#E6E6E6')
plt.rc('lines', linewidth=2)
```

```python
plt.rcParams["figure.figsize"] = (8,5)

resnet1d = [1_113_032, 280_040, 70_904]
computational_time = [390, 187, 121]
test_accuracy = [0.4,0.406,0.344]

fig, ax1 = plt.subplots()

color = 'tab:red'
ax1.set_xlabel('ResNet 1D # params')
ax1.set_ylabel('accuracy', color=color)
ax1.plot(resnet1d, test_accuracy, color=color, label='test accuracy')
ax1.tick_params(axis='y', labelcolor=color)

ax2 = ax1.twinx()  

color = 'tab:blue'
ax2.set_ylabel('time (s)', color=color)  
ax2.plot(resnet1d, computational_time, color=color, label='time spent per epoch')
ax2.tick_params(axis='y', labelcolor=color)

fig.legend(loc="lower right", bbox_to_anchor=(0.95,0.05), bbox_transform=ax1.transAxes, fontsize=16)

plt.show()
```

```python

num_parameters = ["small", "medium", "large"]
# ResNet1D
test_accuracy_model1 = [0.344,0.406,0.441]
computational_time_model1 = [121, 187, 390] 
# ResNet2D
test_accuracy_model2 = [0.528,0.494,0.5]
computational_time_model2 = [79, 153, 208]

# create a 2D array of the accuracy data
data_accuracy = np.array([test_accuracy_model1, test_accuracy_model2])

# create a 2D array of the computational time data
data_time = np.array([computational_time_model1, computational_time_model2])

fig, ax = plt.subplots(figsize=(10, 10))

# plot the computational time data as a heatmap
im = ax.imshow(data_time, cmap='RdYlGn_r')
ax.set_xticks(np.arange(len(num_parameters)))
ax.set_yticks(np.arange(len(data_accuracy)))
ax.set_xticklabels(num_parameters)
ax.set_yticklabels(['ResNet 1D', 'ResNet 2D'])
cbar = ax.figure.colorbar(im, ax=ax, shrink=0.5)
cbar.set_label('time spent per epoch (s)')
plt.setp(ax.get_xticklabels(), rotation=45, ha="right", rotation_mode="anchor")

# add the accuracy data as text inside the heatmap squares
for i in range(len(data_accuracy)):
    for j in range(len(data_accuracy[0])):
        text = ax.text(j, i, '{:.2f}'.format(data_accuracy[i, j]),
                       ha="center", va="center", color="w")

ax.set_title("")

fig.tight_layout()  
plt.show()
```
